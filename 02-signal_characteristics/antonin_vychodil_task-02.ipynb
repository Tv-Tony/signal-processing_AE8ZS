```python
import numpy as np
import matplotlib.pyplot as plt
```

Model: $y(k) = 5\cos(2\pi f_1 k T_s) - 2\sin(2\pi f_2 k T_s)$

```python
# Parameters
fs = 200
N = 100
f1 = 2
f2 = 8
Ts = 1/fs
k = np.arange(N)
```

```python
# Signal
y = 5 * np.cos(2 * np.pi * f1 * k * Ts) - 2 * np.sin(2 * np.pi * f2 * k * Ts)

# Visualization
plt.figure(figsize=(10, 4))
plt.plot(k * Ts, y, 'b-o', markersize=4, label='Sampled Signal y(k)')
plt.title('Discrete Signal y(k)')
plt.xlabel('Time [s]')
plt.ylabel('Amplitude [V]')
plt.grid(True)
plt.legend()
plt.show()
```

### Average Value (Mean) 

$\displaystyle \bar{y} = \frac{1}{N} \sum_{k=1}^{N} y(k)$


```python
manual_mean = sum(y) / N
numpy_mean = np.mean(y)
print(f"Mean: Manual {manual_mean:.4f}, NumPy {np.mean(y):.4f}")
```

### Root Mean Square (RMS)

$\displaystyle RMS = \sqrt{\frac{1}{N} \sum_{k=1}^{N} y^2(k)}$

```python
manual_rms = (sum(y**2) / N)**0.5
numpy_rms = np.sqrt(np.mean(y**2))
print(f"RMS: Manual {manual_rms:.4f}, NumPy {np.sqrt(np.mean(y**2)):.4f}")
```

### Variance (Var) 

$\displaystyle V = \frac{1}{N} \sum_{k=1}^{N} [y(k) - \bar{y}]^2$


```python
manual_var = sum((y - manual_mean)**2) / N
numpy_var = np.var(y)
print(f"Variance: Manual {manual_var:.4f}, NumPy {np.var(y):.4f}")
```

### Standard Deviation (Std) 

$\displaystyle SD = \sqrt{V}$


```python
manual_std = manual_var**0.5
numpy_std = np.std(y)
print(f"Standard Deviation: Manual {manual_std:.4f}, NumPy {np.std(y):.4f}")
```

### Median


```python
sorted_y = sorted(y)
if N % 2 == 0:
    manual_median = (sorted_y[N//2 - 1] + sorted_y[N//2]) / 2
else:
    manual_median = sorted_y[N//2]
numpy_median = np.median(y)
print(f"Median: Manual {manual_median:.4f}, NumPy {np.median(y):.4f}")
```

### Immediate Power

$\displaystyle iP(k) = y^2(k)$

```python
iP = y**2
print(f"Median: Manual {manual_median:.4f}, NumPy {np.median(y):.4f}")
```

### Average Power

$\displaystyle P = \frac{1}{N} \sum_{k=1}^{N} y^2(k)$

```python

manual_avg_power = sum(y**2) / N
numpy_avg_power = np.mean(y**2)
print(f"Average Power: Manual {manual_avg_power:.4f}, NumPy {np.mean(y**2):.4f}")
```

### Energy

$\displaystyle E = \sum_{k=1}^{N} y^2(k)$


```python
manual_energy = sum(y**2)
numpy_energy = np.sum(y**2)
print(f"Energy: Manual {manual_energy:.4f}, NumPy {np.sum(y**2):.4f}")
```

### Instant Power Plotted

```python

# Visualization
plt.figure(figsize=(10, 6))
plt.plot(k * Ts, iP, 'r-o', markersize=4, markerfacecolor='none', markeredgecolor='r')
plt.title('Instant power')
plt.xlabel('t [s]')
plt.ylabel('Instant power [$V^2$]')
plt.axis([0, 0.5, 0, 45])
plt.grid(False) 
plt.show()
```

#### Graph Comarison

```python
data = [
    ("Mean value", manual_mean, numpy_mean, "V"),
    ("RMS value", manual_rms, numpy_rms, "V"),
    ("Variance", manual_var, numpy_var, "V^2"),
    ("Standard deviation", manual_std, numpy_std, "V"),
    ("Median", manual_median, numpy_median, "V"),
    ("Mean power (DSP)", manual_avg_power, numpy_avg_power, "V^2"),
    ("Energy (DSP)", manual_energy, numpy_energy, "V^2")
]

for label, own, mat, unit in data:
    print(f"{label:<20} | {own:<15.4f} | {mat:<15.4f} | {unit}")
```
