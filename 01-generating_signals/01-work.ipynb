```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
```

# Discrete Sinusoidal Signal
The signal is based of the formula signal follows the mathematical formula:
$y(t) = DC + Amp \cdot \sin(2 \pi \cdot F_{sig} \cdot t + fi_0)$

THe signal parameters are adjustable below

```python
Amp = 2.0       # Amplitude [V]
DC = 1.0        # DC offset [V]
fi0 = 0.0       # Initial phase [rad]
F_sig = 0.2     # Signal frequency [Hz]
F_S = 5.0       # Sampling frequency [Hz]

max_time = 10.0 # Duration of the plot [s]
```

```python
# Time Vector
t = np.arange(0, max_time + (1/F_S), 1/F_S) 

# Signal Model
y = DC + Amp * np.sin(2 * np.pi * F_sig * t + fi0)


plt.figure(figsize=(10, 4))
markerline, stemlines, baseline = plt.stem(t, y, linefmt='r-', basefmt=' ', markerfmt='ro')
plt.setp(markerline, markerfacecolor='white', markeredgecolor='red')
plt.setp(stemlines, color='red', linewidth=1.2)

# Formating
plt.axhline(0, color='black', linewidth=1) 
for y_grid in [-1, 1, 2]:                  
    plt.axhline(y_grid, color='gray', linestyle=':', linewidth=1)

plt.title(f"Sinusoidal sampled signal\nAmp= {Amp} V, DC= {DC} V, fi0= {fi0} rad, $F_{{sig}}$ = {F_sig} Hz, $F_S$ = {F_S} Hz", pad=15, fontweight='bold')
plt.xlabel("Time [s]", fontsize=11)
plt.ylabel("Voltage [V]", fontsize=11)

plt.xlim(-0.2, max_time + 0.2)
plt.ylim(-1.5, 3.5)

plt.xticks(np.arange(0, 11, 1))
plt.yticks(np.arange(-1, 4, 1))

plt.tick_params(direction='in')

plt.tight_layout()
plt.show()
```

# Discrete Squared Signal
The signal is based of the model: 
$y(t) = DC + Amp \cdot \text{square}(2 \pi \cdot F_{sig} \cdot t + fi_0)$

Parameters are adjustable below to change the amplitude, frequency, and sampling rate.

```python
Amp = 2.0       # Amplitude [V]
DC = 1.0        # DC offset [V]
fi0 = 0.0       # Initial phase [rad]
F_sig = 0.2     # Signal frequency [Hz]
F_S = 5.0       # Sampling frequency [Hz]

max_time = 10.0 # Duration of the plot [s]
```

```python
# Time Vector
t = np.arange(0, max_time + (1/F_S), 1/F_S) 

# Signnal Model
y = DC + Amp * signal.square(2 * np.pi * F_sig * t + fi0)

# Formating
plt.figure(figsize=(10, 4))
markerline, stemlines, baseline = plt.stem(t, y, linefmt='b-', basefmt=' ', markerfmt='bo')
plt.setp(markerline, markerfacecolor='white', markeredgecolor='blue')
plt.setp(stemlines, color='blue', linewidth=1.2)

plt.axhline(0, color='black', linewidth=1) 
for y_grid in [-1, 1, 2]:                 
    plt.axhline(y_grid, color='gray', linestyle=':', linewidth=1)

plt.title(f"Squared sampled signal\nAmp= {Amp} V, DC= {DC} V, fi0= {fi0} rad, $F_{{sig}}$ = {F_sig} Hz, $F_S$ = {F_S} Hz", pad=15, fontweight='bold')
plt.xlabel("Time [s]", fontsize=11)
plt.ylabel("Voltage [V]", fontsize=11)

plt.xlim(-0.2, max_time + 0.2)
plt.ylim(-1.5, 3.5)

plt.xticks(np.arange(0, 11, 1))
plt.yticks(np.arange(-1, 4, 1))

plt.tick_params(direction='in')

plt.tight_layout()
plt.show()
```

# Discrete Saw-toothed Signal
Signal is based on the mathematical formula:
$y(t) = DC + Amp \cdot \text{sawtooth}(2 \pi \cdot F_{sig} \cdot t + fi_0)$

Parameters are adjustable below. 

```python
Amp = 2.0       # Amplitude [V]
DC = 1.0        # DC offset [V]
fi0 = 0.0       # Initial phase [rad]
F_sig = 0.2     # Signal frequency [Hz]
F_S = 5.0       # Sampling frequency [Hz]

max_time = 10.0 # Duration of the plot [s]
```

```python
# Time Vector
t = np.arange(0, max_time + (1/F_S), 1/F_S) 

# Signal Model
y = DC + Amp * signal.sawtooth(2 * np.pi * F_sig * t + fi0)

# Formating
plt.figure(figsize=(10, 4))

markerline, stemlines, baseline = plt.stem(t, y, linefmt='k-', basefmt=' ', markerfmt='ko')

plt.setp(markerline, markerfacecolor='white', markeredgecolor='black')
plt.setp(stemlines, color='black', linewidth=1.2)

plt.axhline(0, color='black', linewidth=1) 
for y_grid in [-1, 3]:                     
    plt.axhline(y_grid, color='gray', linestyle=':', linewidth=1)

plt.title(f"Saw-toothed sampled signal\nAmp= {Amp} V, DC= {DC} V, fi0= {fi0} rad, $F_{{sig}}$ = {F_sig} Hz, $F_S$ = {F_S} Hz", pad=15, fontweight='bold')
plt.xlabel("Time [s]", fontsize=11)
plt.ylabel("Voltage [V]", fontsize=11)

plt.xlim(-0.2, max_time + 0.2)
plt.ylim(-1.5, 3.5)

plt.xticks(np.arange(0, 11, 1))
plt.yticks(np.arange(-1, 4, 1))

plt.tick_params(direction='in')

plt.tight_layout()
plt.show()
```
